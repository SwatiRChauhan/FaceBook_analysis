```python
import matplotlib.pyplot as plt
import numpy as num
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('D:/Dataset/pseudo_facebook.csv')
```

```python
df.dropna(how="all",inplace=True)
df.isnull().sum()
```

```python
df.ffill(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(['mobile_likes','mobile_likes_received','www_likes','www_likes_received','tenure','dob_month','dob_day','dob_year'],axis=1,inplace=True)
```

```python
df.describe()
```

```python
bins = [0, 12, 19, 35, 50, 100]  # Define age ranges
labels = ['Children', 'Teenagers', 'Young Adults', 'Middle-aged Adults', 'Older Adults']

# Create a new column for age categories
df['Age Category'] = pd.cut(df['age'], bins=bins, labels=labels)

# Count the occurrences of each age category
category_counts = df['Age Category'].value_counts()

# Plotting with Seaborn (Bar Plot)
sns.barplot(x=category_counts.index, y=category_counts.values, palette='viridis')
plt.xlabel('Age Category')
plt.ylabel('Count')
plt.title('Age Distribution by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(x='gender', y='likes', data=df)
plt.xlabel('Gender')
plt.ylabel('Likes')
plt.title('Analysis based on likes')

plt.show()
```

##  Female Like more frequently than male

```python
sns.barplot(x='gender', y='likes_received', data=df)
plt.xlabel('Gender')
plt.ylabel('Likes Received')
plt.title('Analysis based on likes Received')

plt.show()
```

## Female received more likes than Male

```python
# Gender Distribution Pie Chart
gender_counts = df['gender'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['skyblue', 'pink'], startangle=90, explode=[0.05, 0])
plt.title('Gender Distribution', fontsize=16)
plt.show()
```

## Facebook has more male user than female
