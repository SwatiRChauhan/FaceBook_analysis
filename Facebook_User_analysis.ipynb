# Visualizing the dataset


## Plot Age distrubution

```python
plt.figure(figsize=(8, 5))
sns.histplot(dataset['age'], bins=20, kde=True, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### Most of Facebook User belong to age group 20-40

## Gender-based analysis


```python

plt.figure(figsize=(6, 4))
sns.countplot(data=dataset, x='gender', palette='coolwarm')
plt.title('Gender Distribution')
plt.show()
```

### Facebook more Male User than Female

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='friend_count', y='likes', data=dataset, alpha=0.7)
plt.title('Likes vs. Friend Count')
plt.xlabel('Friend Count')
plt.ylabel('Likes')
plt.show()
```

```python
sns.barplot(x='gender', y='likes', data=dataset)
plt.xlabel('Gender')
plt.ylabel('Likes')
plt.title('Analysis based on likes')

plt.show()
```

### Female Like more frequently than male

```python
sns.barplot(x='gender', y='likes_received', data=dataset)
plt.xlabel('Gender')
plt.ylabel('Likes Received')
plt.title('Analysis based on likes Received')

plt.show()
```

### Female received more likes than Male
